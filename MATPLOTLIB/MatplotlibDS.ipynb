```python
import matplotlib.pyplot as plt
import numpy as np
a=np.linspace(1,4,4)
x=np.sin(a)
plt.plot(a,x,marker='o',ms=10,mec='red',mfc='yellow')
plt.show()
```

```python
#(or) plt.plot(a,x,'<marker><linestyle><color>')
plt.plot(a,x,'o--g')
```

```python
plt.plot(a,x,linestyle='dotted',marker='o',c='red',linewidth=2.0)#(or)ls=":"
```

```python
plt.plot(a,x)
plt.xlabel("X-values")
plt.ylabel("Sine of X values")
plt.title("Sine Graph",loc="left")
plt.grid()
plt.show()
```

```python
plt.plot(a,x)
plt.xlabel("X-values")
plt.ylabel("Sine of X values")
plt.title("Sine Graph",loc="left")
plt.grid(color='green',linestyle='--',linewidth=1.5,axis='y')
plt.show()
```

```python
a=np.linspace(1,4,4)
x=np.sin(a)
y=np.cos(a)
plt.subplot(1,2,1)
plt.plot(a,x)
plt.title("Sine Curve")
plt.subplot(1,2,2)
plt.plot(a,y)
plt.title("Cosine Curve")
plt.suptitle("Sin() vs Cos()")
plt.show()
```

```python

plt.scatter(y,x,cmap='nipy_spectral')
plt.scatter(x,y,label=['S','R','D','V'],color='red',s=[20,15,10,5])
plt.colorbar()
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(y,x,c=np.array(['r','g','b','yellow']))
plt.colorbar()
plt.subplot(1,2,2)
plt.scatter(y,x,cmap='nipy_spectral')
plt.colorbar()
plt.show()
```
